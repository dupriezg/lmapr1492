```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
with MPRester("KslnWabX1xzuAHTI") as m:
    structure = m.get_structure_by_material_id("mp-1008559")
    w = CifWriter(structure)
    w.write_file('mp-1008559.cif')
    
from pymatgen.io.cif import CifParser
parser = CifParser("mp-1008559.cif")
structure = parser.get_structures()[0]

from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
new_s=SpacegroupAnalyzer(structure).get_primitive_standard_structure()
    
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from IPython.display import Image, display
c = XRDCalculator(1.54060, 0.1,)

#get the diffraction diagram of the primitive cell
c.show_plot(new_s)


#get the peak intensity, position(in theta), hkl indices and interplanar distance.
pattern = c.get_pattern(new_s)

#PEAK 1
print('FIRST PEAK :')
twotheta = (pattern.x)[0]
intensity = (pattern.y)[0]
hkls = (pattern.hkls)[0]
d_hkl = (pattern.d_hkls)[0]

print('angle 2thta :',twotheta)
print('intensity :',intensity)
print('hkil bravais-miller indices* :',hkls)
print('interplanar distance d :',d_hkl)
print('')

#PEAK 2
print('SECOND PEAK :')
theta = (pattern.x)[1]
intensity = (pattern.y)[1]
hkls = (pattern.hkls)[1]
d_hkl = (pattern.d_hkls)[1]

print('angle 2thta :',twotheta)
print('intensity :',intensity)
print('hkil bravais-miller indices* :',hkls)
print('interplanar distance d :',d_hkl)
print('')

#PEAK 3
print('THIRD PEAK :')
theta = (pattern.x)[2]
intensity = (pattern.y)[2]
hkls = (pattern.hkls)[2]
d_hkl = (pattern.d_hkls)[2]

print('angle 2thta :',twotheta)
print('intensity :',intensity)
print('hkil bravais-miller indices* :',hkls)
print('interplanar distance d :',d_hkl)
print('')

print('* I used the Bravais-Miller four indices system because we have a hexagonal lattice system. This four indices system allow us to see the permutation symmetries more easily. The i redundant index is placed in thid position in the hkil Bravais-Miller system.')









```
